```python
## Q-1 - c
import numpy as np
 
u = np.array([1, 0, 1])
v = np.array([0, 1, 1])
w = np.array([1, 2, 1])
 
print('Does the equality w = 2v-u  holds? Answer:', np.all(w == 2*v-u ))
```
